'''
    Calculate the compound interest on a principal amount given an annual interest rate,
    compounding frequency, and time period in years.

    Parameters:
    - principal (float): The initial amount of money.
    - annual_interest_rate (float): The annual interest rate as a percentage.
    - compounding_frequency (int): The number of times interest is compounded per year.
    - years (float): The number of years the money is invested for.

    Returns:
    - float: The calculated compound interest amount.
''' 




````python
# функция принимает аргументы и возвращает результат 

def calculate_compound_interest(principal, annual_interest_rate, compounding_frequency, years):
    # переведем введенное значение процентов в десятичную дробь (float)
    annual_interest_rate_decimal = annual_interest_rate / 100
    # для лучшей читаемости выведем степень
    compounding_times = compounding_frequency * years

    # вычислим future value вклада
    compound_interest = principal * (1 + annual_interest_rate_decimal / compounding_frequency) ** compounding_times 

    return compound_interest
````

````python
principal_amount = float(input("Enter the principal amount: $"))
interest_rate = float(input("Enter the annual interest rate (as a percentage): "))
compounding_frequency = int(input("Enter the number of times interest is compounded per year: "))
years = float(input("Enter the number of years: "))

result = calculate_compound_interest(principal_amount, interest_rate, compounding_frequency, years)
print(f"Final Amount: ${result:.2f}")
````

В Python существуют общепринятые правила по именованию переменных, описанные в PEP 8. 

1. Имена переменных должны содержать только **маленькие буквы**.
2. Слова в имени переменной **разделяются символами подчеркиваниями** (snake_case). Например, `my_variable`.
3. Имя переменной должно быть описательным и понятным, чтобы улучшить читаемость кода.
4. Используйте английские слова для имен переменных и избегайте использования нелатинских символов.
5. Избегайте использования зарезервированных слов Python (например, `if`, `while`, `for`, `class`) в качестве имен переменных.
6. Имена переменных не должны начинаться с цифры, но могут содержать цифры внутри имени (например, `var1`, `value_2`).
7. Избегайте использования односимвольных имен переменных, за исключением случаев, когда они являются стандартными сокращениями, такими как `i`, `j`, `k` для индексов цикла.
8. Избегайте использования длинных имен переменных, которые усложняют чтение кода.
9. Для константных значений, используйте заглавные буквы и разделяйте слова символами подчеркивания (например, `MAX_VALUE`).

Про форматирование десятичных дробей

1. **Указание количества знаков после запятой:**
   - Используйте `:.nf`, чтобы отобразить число с плавающей точкой с точно `n` знаками после запятой. Например: `{pi:.2f}`.

2. **Заполнение ведущими нулями:**
   - Вы можете заполнить число нулями ведущими нулями с использованием `:0nf`. Например, чтобы отобразить число с 5 знаками после запятой и ведущими нулями: `{pi:07.5f}`.

3. **Минимальная ширина и выравнивание:**
   - Управляйте минимальной шириной вывода с `{value:width.precisionf}`. Также можно указать выравнивание с `<`, `^` или `>`. Например, чтобы отцентрировать число с плавающей точкой в ширине 10 символов: `{pi:^10.2f}`.

4. **Обработка знака:**
   - Можно включить знак числа (положительное или отрицательное) с помощью опций `+` или `-`. Например: `{pi:+.2f}`.

5. **Научная нотация:**
   - Используйте спецификатор формата `e` для научной нотации. Например: `{pi:.2e}`.

6. **Форматирование в соответствии с локальными настройками:**
   - Вы можете применить форматирование, зависящее от локали, с помощью `,`. Например, чтобы использовать запятую в качестве разделителя тысяч: `{large_number:,.2f}`.

7. **Формат процентов:**
   - Чтобы отобразить число с плавающей точкой как процент, умножьте его на 100 и используйте формат `%`. Например: `{percentage:.1%}`.

8. **Пользовательское форматирование с шириной и выравниванием:**
   - Можно объединить ширину, выравнивание и другие параметры форматирования для получения пользовательского форматирования. Например: `{value:*>10,.2f}` для правого выравнивания числа с 2 знаками после запятой в ширине 10 символов с заполнением звездочками.


````python

# Вот пример, демонстрирующий некоторые из этих возможностей:

amount = 12345.6789

formatted_amount = f"Сумма: {amount:*<15,.2f}"  # Правое выравнивание, ширина 15, 2 знака после запятой, разделитель тысяч
print(formatted_amount)

percentage = 0.4567
formatted_percentage = f"Процент: {percentage:.1%}"  # Формат процента с 1 знаком после запятой
print(formatted_percentage)

````

````python
# Input the number of minutes
n = int(input("Enter the number of minutes: "))

# Calculate the number of hours and minutes
hours = (n // 60) % 24  # Number of hours, % 24 to ensure it stays within a 24-hour range
minutes = n % 60       # Number of remaining minutes

# Output the result
print(f"The digital clock will display {hours} hours and {minutes} minutes.")
````

#### Напишем простенькую функцию, чтобы понять код со сложным процентом

Как мы помним, основная концепция в том, что функция получает аргументы и возвращает результат

````python
def square(x):
    x_pow = x ** 2 
    return x_pow
````

````python
x = 2
x_sq = square(x)
print(x_sq)
````
